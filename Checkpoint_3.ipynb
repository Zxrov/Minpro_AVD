# **A.Pengecekan Terhadap Data**

**1. nilai yang hilang (Missing Values)**

```python
print((df.isna().sum() / len(df)) * 100)
```

**2. Duplicated Values**

```python
duplicate_rows = df[df.duplicated()]
print("\nJumlah Baris Duplikat:", len(duplicate_rows))
```

**3. Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/student_performance_data.csv")

df
```

# **1. Visualisasi Data yang Relevan dengan Business Understanding**

```python
print(df.columns)
```

**1. Hubungan antara jam belajar dan GPA.**


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["StudyHoursPerWeek"], y=df["GPA"], alpha=0.6)
plt.xlabel("Jam Belajar per Minggu")
plt.ylabel("GPA")
plt.title("Hubungan antara Jam Belajar dan GPA")
plt.grid(True)
plt.show()
```

**2. Hubungan antara tingkat kehadiran dan GPA.**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["AttendanceRate"], y=df["GPA"], alpha=0.6)
plt.xlabel("Tingkat Kehadiran (%)")
plt.ylabel("GPA")
plt.title("Hubungan antara Tingkat Kehadiran dan GPA")
plt.grid(True)
plt.show()
```

**3. Perbandingan GPA berdasarkan pekerjaan paruh waktu.**

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x=df["PartTimeJob"], y=df["GPA"], palette="coolwarm")
plt.xlabel("Pekerjaan Paruh Waktu")
plt.ylabel("GPA")
plt.title("Perbandingan GPA berdasarkan Pekerjaan Paruh Waktu")
plt.show()
```

# **2. Insight dan tindakan yang dapat diambil berdasarkan hasil visualisasi yang dibuat**

**1. Hubungan antara jam belajar dan GPA.**

 * Grafik menunjukkan sebaran antara jam belajar per minggu dan GPA mahasiswa.

 * Secara umum, ada tren positif: semakin banyak jam belajar, semakin tinggi GPA.

*   **Ada beberapa mahasiswa yang belajar lebih banyak tetapi tetap memiliki GPA rendah.** Ini bisa disebabkan oleh metode belajar yang kurang efektif atau faktor lain seperti stres dan kelelahan.
*   Sebaliknya, ada juga mahasiswa yang belajar lebih sedikit tetapi memiliki GPA tinggi, kemungkinan karena cara belajar mereka lebih efisien.

**2. Hubungan antara tingkat kehadiran dan GPA.**

 * Grafik menunjukkan hubungan antara tingkat kehadiran (%) dan GPA mahasiswa.

 * Secara umum, terlihat bahwa mahasiswa dengan **tingkat kehadiran yang lebih tinggi cenderung memiliki GPA lebih tinggi.**

*   ada juga beberapa mahasiswa dengan kehadiran tinggi tetapi GPA rendah, kemungkinan karena metode belajar yang kurang efektif.
*   Sebaliknya, mahasiswa dengan kehadiran rendah umumnya memiliki GPA lebih rendah, yang bisa disebabkan oleh kurangnya pemahaman materi akibat sering absen.



**3. Perbandingan GPA berdasarkan pekerjaan paruh waktu.**

* Grafik ini membandingkan **GPA mahasiswa yang memiliki pekerjaan paruh waktu** dengan yang **tidak bekerja.**
* Median GPA dari kedua kelompok tampak **hampir sama**, menunjukkan bahwa bekerja paruh waktu **tidak selalu berdampak negatif** pada prestasi akademik.
* Namun, sebaran GPA mahasiswa yang bekerja lebih luas, artinya ada beberapa mahasiswa yang bekerja dan tetap memiliki GPA tinggi, tetapi ada juga yang nilainya lebih rendah.
* Ini bisa menunjukkan bahwa dampak pekerjaan paruh waktu tergantung pada **kemampuan mahasiswa dalam mengatur waktu.**
